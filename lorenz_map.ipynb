```python
from numpy import sin, cos
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from matplotlib import animation, rc
rc("animation", html="jshtml")

def lorenz(state, t, sigma=10., beta=8./3, rho=28.0):
    x, y, z = state
    """Compute the time-derivative of a Lorenz system."""
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

```

```python
x0 = [1, 1, 1]  # starting vector
t = np.arange(0, 40, 0.01)  # one thousand time steps
xt = integrate.odeint(lorenz, x0, t)
```

```python
plt.plot(xt[:, 1], xt[:, 2])
plt.plot([np.sqrt(8/3*27), -np.sqrt(8/3*27)], [27, 27], 'x')
plt.show()
```

```python
z = xt[:,2]
max_z = [z for z, zp, zn in zip(z[1:], z, z[2:]) if zp < z and z > zn]
```

```python
z0 = [1, 1, 1]  # starting vector
t = np.arange(0, 500, 0.01)  # one thousand time steps
zt = integrate.odeint(lorenz, x0, t)

z = zt[:,2]
max_z = [z for zp, z, zn in zip(z, z[1:], z[2:]) if zp < z and z > zn]

plt.plot(max_z[:-1], max_z[1:], 'o', ms=1)
plt.show()
```
