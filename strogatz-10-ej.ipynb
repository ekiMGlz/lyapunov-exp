```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as colors
import matplotlib.cm as cm
plt.style.use(["seaborn-pastel", "vscode"])
```

Cobweb de la ecuacion logistica con 0<=r<=4

```python
r = 3.9
x_k = 0.2
f = lambda x : r*x*(1-x)
iterations = 9

x = np.linspace(0, 1, 50)
plt.axis([0, 1, 0, 1])
plt.plot(x, f(x))
plt.plot(x, x, '--')
plt.plot([x_k, x_k], [0, x_k], '--', color="xkcd:pastel pink")
plt.plot(x_k, 0, 'o', color="xkcd:pastel red")

for i in range(iterations):
    x_k1 = f(x_k)
    plt.plot([x_k, x_k],[x_k, x_k1], '--', color="xkcd:pastel pink")
    plt.plot([x_k, x_k1], [x_k1, x_k1], '--', color="xkcd:pastel pink")
    plt.plot(x_k, x_k1, 'x', color="xkcd:pastel orange") 
    x_k = x_k1
```

```python
l = [0.2]
iters = 6
for i in range(iters):
    l.append(f(l[-1]))


plt.axis([0, 1, -1, 1])
plt.hlines(0, 0, 1)
ars = patches.ArrowStyle.Fancy(5, 5)
cmap = cm.get_cmap("Blues")

rad = -1
parity = 1
plt.plot(l[0], 0, 'x', color=cmap(0))
for i in range(iters-1):
    if np.sign(l[i+1] - l[i]) != parity:
        parity *= -1
        rad = -rad
    
    a = patches.FancyArrowPatch([l[i], 0], [l[i+1], 0], connectionstyle=patches.ConnectionStyle.Arc3(rad=rad), 
                                linewidth=0.2, arrowstyle=ars)
    plt.plot(l[i+1], 0, 'x', color=cmap((i+1)/iters))
    plt.gca().add_patch(a)
    
    rad *= -0.8
    
plt.show()
```

```python
ars = patches.ArrowStyle.Fancy(5, 5)
a = patches.FancyArrowPatch([0, 0], [1, 0], connectionstyle=patches.ConnectionStyle.Arc3(rad=-1),
                            arrowstyle=ars, linewidth=0.5)
plt.gca().add_patch(a)
plt.show()
```
